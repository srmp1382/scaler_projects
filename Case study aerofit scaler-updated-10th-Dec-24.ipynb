```python
#Recommendations on positioning:
#1. Clearly, heavy users already go for the KP781 product
#2. However there is no difference anywhere between moderate users and light users, and likewise between the products KP281 and 481
#3. Hence suggest revising the positioning of KP481 for moderate users by building a few more features into them and positioning it as a product for solid performance and good value
#4. We see that more of the educated people buy the KP781 product even at less income, we can position the KP481 position at this bracket
#5. We can link the mileage and wear and tear to the product, like Decathlon does for its shoes or bags, they quantify the usage that the product is meant for
```

```python
data['income_ranges']=pd.qcut(data['Income'],[0,0.3,0.6,1],labels=['Low','Medium','High'])
data['miles_ranges']=pd.qcut(data['Miles'],[0,0.3,0.6,1],labels=['Low','Medium','High'])
```

```python
data['edu_ranges']=pd.qcut(data['Education'],[0,.3,.6,1],labels=['Short education','Medium education','Long education'])
data['age_ranges']=pd.qcut(data['Age'],[0,.3,.6,1], labels=['Young','Middle aged','Senior'])
```

```python
#Overall insights from the case study
#1. The positioning of KP481 needs to be changed to a more rational proposition: 
#It should be targeted at runners who run a little more, and cannot afford a premium 781 treadmill yet.
#Currently it is not differentiated from the KP281 model
#2. The positioning must include the idea of usage and wear and tear, like Decathlon does with its shoes, or like General Motors does with its cars
#   something like: if you are aiming to run upto 2 days a week to start off, then buy the KP281.
#   for 3-4 days a week of moderate exercise, buy the KP481
#   for 5-7 days a week, buy KP781
#3. The company should aim to in sell more 781 or 481 treadmill models based on features other than the amount of running
#   For example, higher speed limit controls, better grips on the handle, and incline adjustment could be one feature in a higher model
#   Other features could be set workouts like interval training etc for advanced runners, which could be on higher models
#4. Even though we know that the amount of education should have nothing to do with buying a treadmill, it appears that people who have more education are more likely to buy a premium 781 treadmill.
#This is a good opportunity to look at people who do not have such a long education and convince them that fitness is directly correlated with the treadmill you buy.
#5. The KP481 is not positioned correctly among male users in terms of usage:
#The only male users are outliers
```
